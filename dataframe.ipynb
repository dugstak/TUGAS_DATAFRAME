```python
import pandas as pd
data_sumber_jabar = "disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.csv"
df = pd.read_csv(data_sumber_jabar)
df
```

```python
pilih_tahun = 2016
jumlah_produksi_sampah = 0
for index, row in df.iterrows():
    if row['tahun'] == pilih_tahun:
        jumlah_produksi_sampah += row['jumlah_produksi_sampah']

print(f"Total Produksi Sampah di Jawa Barat untuk Tahun {pilih_tahun}: {jumlah_produksi_sampah:.3f} ton")
```

```python
data_kota = {}
for index, row in df.iterrows():
    key = (row['nama_kabupaten_kota'], row['tahun'])
    if key in data_kota:
        data_kota[key] += row['jumlah_produksi_sampah']
    else:
        data_kota[key] = row['jumlah_produksi_sampah']

data_kota = pd.DataFrame([
    {"Kabupaten/Kota": k[0], "tahun": k[1], "Total Produksi Sampah (ton)": v}
    for k, v in data_kota.items()
])

print("Total Produksi per Kabupaten/Kota per Tahun:")
data_kota
```

```python
data_kota.to_csv("total_produksi_sampah_kota_per_tahun.csv", index=False)

print("data sudah terexport")
```
